```python
import pprint
import requests
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

## Word Cloud for Apple

```python
url = 'https://newsapi.org/v2/everything?'
secret= 'b118a54dea824fe6865aa81cf93a32fd'
parameters = {
    'q': '$AAPL', # query phrase
    'pageSize': 100,  # maximum is 100
    'apiKey':secret # your own API key
}
```

```python
response = requests.get(url, params=parameters)

# Convert the response to JSON format and pretty print it
response_json = response.json()
```

```python
text_apple = ''
# Loop through all the headlines and add them to 'text_combined' 
for i in response_json['articles']:
    text_apple += i['title'] + ' ' # add a space after every headline, so the first and last words are not glued together
```

```python
wordcloud = WordCloud(max_font_size=50,background_color="black").generate(text_apple)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
