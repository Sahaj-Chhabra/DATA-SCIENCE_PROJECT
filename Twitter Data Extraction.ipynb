```python
# Merging all excel files together
# Boeing.xlxs has all data from 16th March-24th March. Same goes for all other comapnies as well

# filenames
excel_names = ["BA25.xlsx", "BA26.xlsx","BA27.xlsx","BA28.xlsx","Boeing.xlsx"]

# read them in
excels = [pd.ExcelFile(name) for name in excel_names]

# turn them into dataframes
frames = [x.parse(x.sheet_names[0], header=None,index_col=None) for x in excels]

# delete the first row for all frames except the first
# i.e. remove the header row -- assumes it's the first
frames[1:] = [df[1:] for df in frames[1:]]

# concatenate them..
combined = pd.concat(frames)

# write it out
combined.to_excel("Boeing_March.xlsx", header=False, index=False)
```
