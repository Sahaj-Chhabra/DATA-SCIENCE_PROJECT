```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
#sites=[]
#limitation_no=[56,22,22,24,56]
website=[]
headlines=[]
lights=[]
site_aapl = 'http://www.nasdaq.com/symbol/aapl/news-headlines?page='
```

```python
for i in range(1,61):
    website.append(site_aapl+str(i))
    
```

```python
for webpage in website:
    html = requests.get(webpage).content
    soup = BeautifulSoup(html, 'lxml')
    links = soup.find_all('a')
    urls = [link.get('href') for link in links]
    urls = [url for url in urls if url is not None]
    #print(len(urls))
    news_urls = [url for url in urls if '/article/' in url]
    #print(len(news_urls))
    for i in range((len(news_urls)-10),len(news_urls)):
        headlines.append(news_urls[i])
    
```

```python
print(headlines)
```

```python
for head in headlines:
    #head=str(head)
    a = head.replace("https://www.nasdaq.com/article/","")
    print(a)
    b = a.replace("-"," ")
    #b.remove(word.startswith("cm"))
    b=b[:-9]
    print(b)
    lights.append(b)
```

```python
print(lights)
```

```python
d= {"AAPL":lights}
print(d)
```

```python
df =pd.DataFrame(d)
df
```
